# **Movimiento de un Satélite Artificial a partir de las Leyes de Kepler**


A partir de resolver la ecuacion de Kepler para el movimiento:
$$
E - e \sin E = \sqrt{\frac{GM}{a^3}} (t - t_p)
$$

# Condiciones Iniciales: 

\begin{aligned}
    a &= 1.30262 \times R_{\oplus} \\
    e &= 0.16561 \\
    \omega &= 15^\circ \\
    t_p &= 2025 - 03 - 31 \; 00:00:00 \; \textit{UTC}
\end{aligned}



```python
import numpy as np
from astropy.constants import G, M_earth,R_earth
from astropy.time import Time
import matplotlib.pyplot as plt
from astropy.time import TimeDelta


# Definimos los parámetros físicos con Astropy
e =0.16561 # Excentricidad


omega=np.radians(15)
a = 1.30262 * R_earth.value # Semi-eje mayor en radios terrestres

# Definimos los tiempos en formato UTC
t_p = Time("2025-03-31 00:00:00", format="iso", scale="utc")  # Momento del periastro

```

```python


def kepler_equation(a, e, t, t_p, E0=0.0, tol=1e-10, max_iter=100):
    # Convertir tiempos a segundos y calcular la anomalía media M
    delta_t = (t - t_p).to_value('sec')
    M = np.sqrt(G.value * M_earth.value / a**3) * delta_t  # Anomalía media

    # Definimos la ecuación de Kepler y su derivada
    def f(E):
        return E - e * np.sin(E) - M

    def df(E):
        return 1 - e * np.cos(E)
    
    E = E0  # Inicialización con el valor inicial

    for _ in range(max_iter):
        fE = f(E)
        dfE = df(E)
        
        if abs(dfE) < 1e-12:  # Evita divisiones por cero
            raise ValueError("Derivada cercana a cero. Método falló.")
        
        E_new = E - fE / dfE  # Método de Newton-Raphson
        
        if abs(E_new - E) < tol:  # Criterio de convergencia
            return E_new
        
        E = E_new  # Actualización para la siguiente iteración

    raise ValueError("El método de Newton no convergió después de {} iteraciones.".format(max_iter))

```

```python


def true_anomaly(E, e):
    c = np.sqrt((1 + e) / (1 - e))
    f = 2 * np.arctan(c* np.tan(E / 2))
    return f
```

```python
# Función para calcular la posición
def position(t):
    # Calcular la anomalía media M
    n = np.sqrt(398600.4405 * 1e9 / a**3)  # Movimiento medio en rad/s
    M_t = n * (t - t_p).to('s').value

    # Resolver la ecuación de Kepler para obtener E
    E_t = kepler_equation(a, e, t, t_p)

    # Calcular la anomalía verdadera
    f_t = 2 * np.arctan(np.sqrt((1 + e) / (1 - e)) * np.tan(E_t / 2))

    # Calcular el radio orbital
    r_t = a * (1 - e**2) / (1 + e * np.cos(f_t))

    # Calcular el ángulo polar
    phi_t = f_t + omega

    # Convertir a grados
    phi_t_deg = np.degrees(phi_t)
    f_t_deg = np.degrees(f_t)

    return {
        "Anomalía excéntrica (E)": E_t,
        "Anomalía verdadera (f)": f_t_deg,
        "Radio orbital (r)": r_t,
        "Ángulo polar (phi)": phi_t_deg,
        "x": r_t * np.cos(phi_t),
        "y": r_t * np.sin(phi_t)
    }


```

# **ORBITA DEL SATELITE**

```python

def orbit():
    # Calcular el período orbital en segundos
      # Parámetro gravitacional de la Tierra en m^3/s^2
    T = 2 * np.pi * np.sqrt(a**3 /(G.value * M_earth.value) )  # Período en segundos

    # Crear valores de tiempo hasta un período orbital
    t_values = [t_p + TimeDelta(dt, format='sec') for dt in np.linspace(0, T, 800)]

    # Calcular posiciones
    positions = [position(t) for t in t_values]
    x_values = [p["x"] / 1000 for p in positions]
    y_values = [p["y"] / 1000 for p in positions]

    # Crear la figura
    plt.figure(figsize=(5, 5))  # Tamaño ajustado
    
    # Graficar la órbita hasta T
    plt.plot(x_values, y_values, label="Órbita (hasta T)", color='dodgerblue', linewidth=2)

    # Posición de la Tierra en el centro
    plt.scatter([0], [0], color='red', marker='o', s=80, label="Tierra")

    # Mejoras visuales
    plt.xlabel("x (km)", fontsize=10)
    plt.ylabel("y (km)", fontsize=10)
    plt.title("Órbita del Satélite", fontsize=12, fontweight='bold')
    plt.legend(fontsize=9)
    plt.axis("equal")
    plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)


    plt.axhline(0, color='black', linewidth=0.6, alpha=0.6)
    plt.axvline(0, color='black', linewidth=0.6, alpha=0.6)

    # Mostrar la gráfica
    plt.show()
```

```python
def position_radial(t_values):
# Crear la lista de tiempos
    
    # Calcular posiciones para cada tiempo
    positions = [position(t) for t in t_values]

    # Extraer los valores de r_t
    r_t= [pos["Radio orbital (r)"] for pos in positions]

    return t_values, r_t



```

```python
T = 2 * np.pi * np.sqrt(a**3 /(G.value * M_earth.value) )  # Período en segundos

t_values = [t_p + TimeDelta(dt, format='sec') for dt in np.linspace(0, T, 800)]

t_values_result, r_t_result = position_radial(t_values)


t_seconds = [(t - t_p).to('s').value for t in t_values_result]

# Graficar
plt.plot(t_seconds, r_t_result)
plt.xlabel("Tiempo (s)")
plt.ylabel("Radio orbital (m)")
plt.title("Evolución del Radio Orbital con el Tiempo")
plt.grid()
plt.show()
```

```python

def quadratic_interpolation(x, y, x_target):
    for i in range(len(x) - 1):
        if x[i] <= x_target <= x[i + 1]:
            if i > 0:
                indices = [i - 1, i, i + 1]
            else:
                indices = [i, i + 1, i + 2]
            break
    else:
        return None  # No se encontró un rango válido

    x1, x2, x3 = x[indices[0]], x[indices[1]], x[indices[2]]
    y1, y2, y3 = y[indices[0]], y[indices[1]], y[indices[2]]

    return (
        y1 * ((x_target - x2) * (x_target - x3)) / ((x1 - x2) * (x1 - x3)) +
        y2 * ((x_target - x1) * (x_target - x3)) / ((x2 - x1) * (x2 - x3)) +
        y3 * ((x_target - x1) * (x_target - x2)) / ((x3 - x1) * (x3 - x2))
    )

def find_time(r0, t_values, r_values):
    t_seconds = np.array([(t - t_p).to('s').value for t in t_values])
    r_values = np.array(r_values)

    if not (r_values.min() <= r0 <= r_values.max()):
        return None, None, None

    # Encuentra el tiempo estimado en el que se alcanza r0
    t0 = quadratic_interpolation(r_values, t_seconds, r0)

    # Interpola el valor de r en t0
    r_interpolado = quadratic_interpolation(t_seconds, r_values, t0)

    # Diferencia entre el valor interpolado y el real
    error_r = abs(r_interpolado - r0)

    return t_p + TimeDelta(t0, format='sec'), r_interpolado, error_r

# Parámetro radial
r0 = 1.5 * R_earth.value  # 1.5 veces el radio de la Tierra en metros


t0, r_interp, error_r = find_time(r0, t_values_result, r_t_result)

if t0:
    print(f"La distancia radial r0 = {r0/1e3:.6f} km es alcanzada en la época {t0.iso}")
    print(f"Valor interpolado de r en t0: {r_interp/1e3:.6f} km")
    print(f"Diferencia |r_interpolado - r0|: {error_r:.6f} m")
else:
    print("No se encuentra en la órbita.")
```
